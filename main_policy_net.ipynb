```python
import numpy as np
import gym
import random
from pathlib import Path

from Source import PolicyNetwork
```

```python
path = Path()
d = path.resolve()
weight_save_path = str(d) + "/weights/policy_net.ckpt"
weight_load_path = str(d) + "/weights/policy_net.ckpt"
```

```python
env = gym.make('CartPole-v0')
model = PolicyNetwork(env = env)
model.train(n_epochs = 0, weight_load_path = weight_load_path)
```

```python
timesteps = 1000
s = env.reset()
action_list = []
for t in range(timesteps):
    env.render()
    action = model.act(s)
    action_list.append(action)
    print("Action at time step " + str(t) + ": " + str(action))
    if t > 5 and action_list[-1] == action_list[-2] == action_list[-3] == action_list[-4]:
        print(env.state)
    s, reward, done, info = env.step(action)
    if done:
        print("Episode finished after {} timesteps".format(t+1))
        break
```
