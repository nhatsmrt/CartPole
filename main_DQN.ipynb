```python
import numpy as np
import gym
import random

from Source import DQN
```

```python
env = gym.make('CartPole-v0')
model = DQN(env = env)
model.train(n_epochs = 1000)
```

```python
timesteps = 100
s = env.reset()
for t in range(timesteps):
    env.render()
    action = model.act(s)
    print("Action at time step " + str(t) + ": " + str(action))
    s, reward, done, info = env.step(action)
    if done:
        print("Episode finished after {} timesteps".format(t+1))
        break
```

```python
import tensorflow as tf
from pathlib import Path


saver = tf.train.Saver()
d = Path().resolve()
weight_save_path = str(d) + "/weights/DQN.ckpt"
saver.save(DQN._sess, save_path = weight_save_path)
```
